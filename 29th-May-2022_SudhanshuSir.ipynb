```python
# q1 : Try to print this by using while loop
n=9
i = 1
while i <= n :
    j = 1
    while j <= i:
        print("*", end = " ")
        j += 1
    print()
    i += 1
```

```python
# q15 :write a fun which will be able to print an index of all premitive element which you will pass.
u=1
n=input('Enter Max Elment you want to insert in Store value :')

MyElements=[]
def Storevalue(MyElements,PremitiveElement):
    MyElements.append(PremitiveElement)
while u <= int(n):
    v=input('Enter Values to Store : ')
    Storevalue(MyElements,v)
    u+=1
print(MyElements)
search_index=input('Enter the Value you want to find out the index :')
if search_index in MyElements:
    result = MyElements.index(search_index)
    print(f"The {search_index} has an index of {result}.")
else:
    print(f"{search_index} doesn't exist in the list.")

```
